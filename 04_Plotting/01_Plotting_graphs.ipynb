## Scatter Plot

Specify that you want a scatter-plot with the **kind=** argument:  

`kind = scatter`  

A scatter plot needs an a x-axis and a y-axis.  

In the example below we will use "Duration" for the x-axis and "Calories" for the y-axis.  

Include the x and y arguments like this:

x = 'Duration', y = 'Calories'


#### Example: Plotting a scatter graph

```python
#plotting using kind argumnet and specifying the asix
df.plot(kind = 'scatter', x = 'Duration', y = 'Calories')

plt.show()
```

We can derive from the graph that 'Duration' and 'Calories' have higher correlation (`0.922721`), as   
we concluded in the previous chapter.

Let's create a scatterplot, where there is bad relationship between columns, like 'Duration' and  
'Maxpulse', with correlation `0.009403`.

#### Example: Plotting Scatterplot of 'Duration' and 'Maxpulse'
